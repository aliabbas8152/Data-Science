### HTTP Methods in Flask:

```python
from flask import Flask, request

app = Flask(__name__)

@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'GET':
        return 'This is a GET request.'
    elif request.method == 'POST':
        return 'This is a POST request.'
    else:
        return 'Unsupported request method.'

if __name__ == '__main__':
    app.run(debug=True)
```

### Templates in Flask:

1. Create a `templates` folder in your project directory.

2. Inside the `templates` folder, create an HTML file, e.g., `index.html`:

   ```html
   <!DOCTYPE html>
   <html lang="en">
   <head>
       <meta charset="UTF-8">
       <meta name="viewport" content="width=device-width, initial-scale=1.0">
       <title>{{ title }}</title>
   </head>
   <body>
       <h1>{{ message }}</h1>
   </body>
   </html>
   ```

3. Modify your Flask app to render this template:

   ```python
   from flask import Flask, render_template

   app = Flask(__name__)

   @app.route('/')
   def index():
       return render_template('index.html', title='Flask Template', message='Hello, Flask!')

   if __name__ == '__main__':
       app.run(debug=True)
   ```
