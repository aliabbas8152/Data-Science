### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```
#### Note: Also add title, xlabel, and ylabel to the plot.

**A2: A scatter plot** is a visualization technique used to represent relationships between two independent variables. Each data point in a scatter plot corresponds to a pair of values (x, y), and is typically marked with a symbol like a circle, square, or triangle.

Here's the code to generate and plot a scatter plot using the provided data:


````python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for x and y (using provided code)
np.random.seed(3)  # Set seed for reproducibility (optional)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Customize the size of the plot
plt.figure(figsize=(6,3))

# Create the scatter plot
plt.scatter(x, y)

# Customize the plot
plt.title('Representation of Scatter Plot')
plt.xlabel('This is X-axis')
plt.ylabel('This is Y-axis')

# Show the plot
plt.grid(True)  # Add grid lines for better readability
plt.show()
````


**Explanation:**

1. **Import Libraries:** We import `matplotlib.pyplot` as `plt` for plotting and `numpy` as `np` for random number generation.
2. **Generate Data:** The code snippet generates 50 random data points for x and y with a mean of 3 and a standard deviation of 2. Setting a seed ensures the same data is generated each time you run the code (optional).
3. **Create Scatter Plot:** `plt.scatter(x, y)` creates the scatter plot, plotting each data point (x, y) with a default marker style (circle).
4. **Customize Plot:** We add a title, x-axis label, and y-axis label for clarity using `plt.title`, `plt.xlabel`, and `plt.ylabel` functions.
5. **Grid Lines (Optional):** `plt.grid(True)` adds grid lines for easier visual reference.
6. **Show Plot:** `plt.show()` displays the generated scatter plot.

This code will create a scatter plot with 50 randomly generated data points, each marked with a circle. You can experiment with different marker styles and colors using the `marker` and `c` arguments in the `plt.scatter` function.

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
```python
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

**A3: Why use subplot()**

The `subplot()` function in Matplotlib's `pyplot` module is used to create multiple subplots within a single figure. This allows you to efficiently visualize and compare multiple datasets or relationships on the same canvas. It's particularly useful when you want to see trends or patterns across different data series simultaneously.

Here's the code to create four line plots using `subplot()`:

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for each line
line1_x = np.array([0, 1, 2, 3, 4, 5])
line1_y = np.array([0, 100, 200, 300, 400, 500])

line2_x = np.array([0, 1, 2, 3, 4, 5])
line2_y = np.array([50, 20, 40, 20, 60, 70])

line3_x = np.array([0, 1, 2, 3, 4, 5])
line3_y = np.array([10, 20, 30, 40, 50, 60])

line4_x = np.array([0, 1, 2, 3, 4, 5])
line4_y = np.array([200, 350, 250, 550, 450, 150])

# Create the figure and subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # Adjust figsize for better visualization

# Plot each line on a separate subplot
axes[0, 0].plot(line1_x, line1_y, label='Line 1')  # Top-left subplot
axes[0, 0].set_title('Line 1 Data')
axes[0, 0].legend()

axes[0, 1].plot(line2_x, line2_y, label='Line 2')  # Top-right subplot
axes[0, 1].set_title('Line 2 Data')
axes[0, 1].legend()

axes[1, 0].plot(line3_x, line3_y, label='Line 3')  # Bottom-left subplot
axes[1, 0].set_title('Line 3 Data')
axes[1, 0].legend()

axes[1, 1].plot(line4_x, line4_y, label='Line 4')  # Bottom-right subplot
axes[1, 1].set_title('Line 4 Data')
axes[1, 1].legend()

# Customize the overall plot (optional)
fig.suptitle('Four Line Plots', fontsize=12)  # Add a common title for the entire figure
plt.tight_layout()  # Adjust spacing between subplots for better readability

# Show the plot
plt.show()
````

**Explanation:**

1. **Import Libraries:** We import `matplotlib.pyplot` and `numpy`.
2. **Define Data:** We define the x and y data for each line plot using NumPy arrays.
3. **Create Subplots:** `plt.subplots(2, 2, figsize=(10, 6))` creates a figure with a grid of 2 rows and 2 columns (4 subplots) and sets the figure size for better visualization (optional).
4. **Plot Lines:** We use a loop or indexing to iterate over each line's data and plot them on separate subplots using `axes[row, col].plot(x, y, label='Line Label')`. We also set titles and legends for each subplot.
5. **Optional Customization:** We add a common title for the entire figure using `fig.suptitle` and adjust spacing between subplots using `plt.tight_layout`.
6. **Show Plot:** `plt.show()` displays the generated plot with four subplots.

Note: This code creates a single figure containing four subplots, each displaying a different line plot with its corresponding data. You can customize the plot further by adding labels, adjusting colors, or changing line styles.

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

**A4:** A bar plot is a graphical representation of categorical data using rectangular bars. The length or height of each bar corresponds to the value it represents for a particular category. Bar plots are commonly used for:

* **Comparing Values:** They are effective for visually comparing values between different categories.
* **Displaying Frequencies:** They can represent frequencies or counts of data points within each category.
* **Presenting Proportions:** When bar heights are normalized (total adds up to 1), they can show proportions or percentages.

**Code for Bar Plot:**

````python
import matplotlib.pyplot as plt
import numpy as np

# Define company names and profits
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(6, 4))  # Set figure size for better readability
plt.bar(company, profit, color=['red', 'green', 'blue', 'purple'])  # Set custom colors
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company Profits (Vertical Bar Plot)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show plot
plt.show()
````

**Code for Horizontal Bar Plot:**

````python
# Create a horizontal bar plot
plt.figure(figsize=(6, 4))  # Set figure size for horizontal plot
plt.barh(company, profit, color=['red', 'green', 'blue', 'purple'])  # Use 'barh' for horizontal bars
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.gca().invert_yaxis()  # Invert y-axis to display highest profit at the top

# Show plot
plt.show()
````

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

**A5:** A box plot is a visual representation of the distribution of a dataset. It summarizes key statistical information:

* **Center:** The median (middle value) of the data.
* **Spread:** The interquartile range (IQR) representing the middle 50% of the data. The box extends from the first quartile (Q1) to the third quartile (Q3).
* **Whiskers:** Extend to data points within 1.5 times the IQR from the quartiles. Values beyond this range are considered outliers and are plotted individually.

Box plots are useful for:

* **Comparing Distributions:** They allow you to compare the distribution of data between different groups or categories.
* **Identifying Outliers:** You can easily identify potential outliers that lie outside the whiskers of the box.
* **Assessing Skewness:** The symmetry or skewness of the data distribution can be visually observed based on the box shape.

**Code for Box Plot:**


````python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for two boxes
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create the box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'], notch=True, vert=True, patch_artist=True)
plt.xlabel('Data Box')
plt.ylabel('Values')
plt.title('Box Plot of Two Datasets')

# Show the plot
plt.grid(True)
plt.show()
````
