# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
# create a pair of x and y 
plt.scatter(x,y)
```

```python
# c is for datapoints' color, alpha is for datapoints' intensity, marker is for to choose type of datapoints' shape,  

plt.figure(figsize=(6,2))      # to change the size of the graph
plt.scatter(x,y, c='#223BDC', alpha=.9, marker='X')
plt.xlabel("This is x-axis")   # to label x-axis
plt.ylabel("This is y-axis")   # to label y-axis
plt.title("x vs y graph")      # to put title to the graph
plt.grid()                    # to put grids in the graph
```

```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y)
```

```python
# Creating Vertical Bar Graph
plt.figure(figsize=(5,3))
plt.bar(x,y, color='r')
plt.title("This is Bar Plot")
plt.xlabel("Representing my Categorical Data")
plt.ylabel("Representing my Numerical Data")
plt.grid()
```

```python
# Creating Horizontal Bar Graph
plt.figure(figsize=(5,3))
plt.barh(x,y, color='b')
plt.title("This is Bar Plot")
plt.xlabel("Representing my Categorical Data")
plt.ylabel("Representing my Numerical Data")
plt.grid()
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
# creating the histogram graph to show the repetation of the data
plt.figure(figsize=(6,3))
plt.hist(data, color = '#700EBE')
plt.title("This is Histogram Graph")
plt.xlabel("Data(Unique)")
plt.ylabel("Frequency of Data")
plt.grid()
plt.savefig("Test.png")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c='r', alpha = .8)
plt.title("This is a 3D Graph")
plt.xlabel("This is x-axis")
plt.ylabel("This is y-axis")
ax.set_zlabel('Z-axis')
plt.show()
```
