### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create a figure to arrange two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))  # Adjust figsize for better visualization

# First box plot: age vs pclass
sns.boxplot(
    x = "pclass",
    y = "age",
    showmeans=True,  # Display means as points within the boxes
    data=titanic_data,
    ax=ax1
)
ax1.set_title("Age Distribution by Passenger Class")
ax1.set_xlabel("Passenger Class")
ax1.set_ylabel("Age")

# Second box plot: fare vs pclass
sns.boxplot(
    x = "pclass",
    y = "fare",
    showmeans=True,
    data=titanic_data,
    ax=ax2
)
ax2.set_title("Fare Distribution by Passenger Class")
ax2.set_xlabel("Passenger Class")
ax2.set_ylabel("Fare")

# Adjust layout (optional)
plt.tight_layout()  # Adjust spacing to prevent overlapping elements

# Display the plot
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for 'cut'
sns.histplot(
    x="price",
    hue="cut",
    data=diamonds_data,
    stat="density",  # Use density estimation for smoother distribution representation
    common_norm=False,  # Avoid normalizing histograms to have the same total area
    palette="Set1"  # Choose a color palette suitable for multiple categories
)

# Add title and labels (optional)
plt.title("Distribution of Diamond Price by Cut")
plt.xlabel("Price (USD)")
plt.ylabel("Density")

# Show the plot
plt.show()
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import pandas as pd

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for 'species'
sns.pairplot(iris_data, hue="species", diag_kind="hist")  # Use histograms for diagonal plots

# Add a title (optional)
plt.suptitle('Pairwise Relationships in Iris Dataset Colored by Species')

# Show the plot
plt.show()
```

**Explanation:**

1. **Import Libraries:** We import `seaborn` for data loading and plotting and `pandas` (although not strictly necessary here) for potential data manipulation with DataFrames.
2. **Load Dataset:** We use `sns.load_dataset("iris")` to load the built-in "iris" dataset.
3. **Pair Plot:** We use `sns.pairplot(iris_data, hue="species")` to create the pair plot.
   - `iris_data` is the DataFrame containing the iris dataset.
   - `hue="species"` uses the 'species' column to color-code the scatter plots based on the iris flower species.
4. **Diagonal Plots:** We set `diag_kind="hist"` to display histograms on the diagonal plots of the pair grid. This allows you to see the distribution of each feature within each species. 
5. **Title (Optional):** We add a title using `plt.suptitle` for better context.
6. **Show Plot:** We use `plt.show()` to display the final pair plot with scatter plots colored by species and histograms for each feature on the diagonal.

This code effectively utilizes Seaborn's `pairplot` function with the `hue` parameter to create a comprehensive visualization that helps you explore the relationships between all pairs of features in the iris dataset while highlighting potential differences based on the flower species.

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to prepare it for heatmap
flights_pivot = flights_data.pivot_table(index="month", columns="year", values="passengers")

# Plot a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passenger Numbers by Year and Month")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

**Explanation:**
- We use the `pivot_table()` function instead of `pivot()` to pivot the dataset. This function allows us to specify the index, columns, and values separately.
- We specify the "month" column as the index, the "year" column as the columns, and the "passengers" column as the values.
- The rest of the code remains the same.
