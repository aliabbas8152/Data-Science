### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

**Ans-1:** The five plots that you can create using the Seaborn library, along with their uses:

1. **Distribution Plot (Histogram or Kernel Density Plot):**

   * **Use:** Visualize the distribution of a single variable. 
   * **Helpful for:** Understanding the shape of the data (normal, skewed, etc.), identifying outliers, and comparing distributions across groups.

2. **Scatter Plot:**

   * **Use:** Explore the relationship between two continuous variables.
   * **Helpful for:** Identifying trends, correlations, and potential non-linear relationships between variables.

3. **Bar Plot:**

   * **Use:** Compare values across different categories of a single variable.
   * **Helpful for:** Visualizing group means, proportions, or other statistical summaries.

4. **Box Plot:**

   * **Use:** Compare distributions of a variable across different categories.
   * **Helpful for:** Identifying differences in medians, interquartile ranges (spread), and potential outliers between groups.

5. **Heatmap:**

   * **Use:** Visualize relationships between multiple categorical variables or represent a matrix as color gradients.
   * **Helpful for:** Exploring correlations between categorical variables, identifying patterns in large datasets, and visualizing complex relationships.


### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

#### Note: timepoint, signal, event, and region are columns in the fmri dataset.

**Ans-2:**

```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")
```

```python
fmri_data
```

**Line plots for each event**

```python
# Plot line plots for each event
g = sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri_data, height=5, aspect=2)
g.figure.suptitle('Line Plot of Signal vs Timepoint for Different Events', y=1.05)
plt.xlabel('Timepoint')
plt.ylabel('Signal')

plt.show()
```

**Line plots for each region**

```python
# Plot line plots for each region
g = sns.relplot(x="timepoint", y="signal", hue="region", kind="line", data=fmri_data, height=5, aspect=2)
g.figure.suptitle('Line Plot of Signal vs Timepoint for Different Regions', y=1.05)
plt.xlabel('Timepoint')
plt.ylabel('Signal')

plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

**Ans-6:**

```python
# import library
import seaborn as sns
```

```python
# Load the 'flights' dataset
flights_data = sns.load_dataset('flights')
```

```python
flights_data
```

```python
# Pivot the dataset to prepare it for heatmap
flights_pivot = flights_data.pivot_table(index="month", columns="year", values="passengers")
```

**Heatmap plot for the 'flights' dataset:**

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot)
```
